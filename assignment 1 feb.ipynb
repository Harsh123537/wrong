```python
# We use for loop when we know that how many times we have to iterate the loop
print("example of for loop")
l=[3,5,6,7,8,9]
for i in range(len(l)):
    print(i)
#Here we know that we have to iterate the given list till the length of list.

#We use while loop when we don't know that how many times we have to iterate the loop we know the terminating condition.
print()
print("example of while loop")

num=1000
cnt=0
while num>0:
    num=num//5
    cnt=cnt+1
print(cnt)
#Here we have not a knowledge that hoe many times the loop will run we have only terminating condition.
```

```python
n=10
i=1
sum1=0
pro1=1
while i<=n:
    sum1=sum1+i
    pro1=pro1*i
    i=i+1
print(sum1)
print(pro1)
sum2=0
pro2=1
for i in range(1,n+1):
    sum2=sum2+i
    pro2=pro2*i
print(sum2)
print(pro2) 
```
